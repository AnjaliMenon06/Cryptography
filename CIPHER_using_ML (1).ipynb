**Generating Cipher Text**

```python
def generateKey(string, key):
    key = list(key)
    if len(string) == len(key):
        return(key)
    else:
        for i in range(len(string) -
                       len(key)):
            key.append(key[i % len(key)])
    return("" . join(key))
     
def cipherText(string, key):
    cipher_text = []
    for i in range(len(string)):
        x = (ord(string[i]) +
             ord(key[i])) % 26
        x += ord('A')
        cipher_text.append(chr(x))
    return("" . join(cipher_text))

string = plaintext
keyword1 = "thechroniclesofnarnia"
keyword2 = "orderofthephoenix"
key1 = generateKey(string, keyword1)
key2 = generateKey(string, keyword2)
cipher_text_correct = cipherText(string,key1)   
cipher_text_incorrect = cipherText(string,key2)    
```

**Converting the cipher text and plain text  to ascii then to 8-bit binary value**

```python
def convertion(text):
  ascii_value =[]
  binary_str=[]
  
  for i in range(len(text)):
    ascii_value.append(ord(text[i]))
  
  for i in range(len(ascii_value)):
    binary_value = bin(ascii_value[i])[2:]
    while len(binary_value) < 8:
      binary_value += '0'
      bnr = binary_value[::-1]
      binary_str.append(bnr)
  return binary_str

binary_ciphertext_1=convertion(cipher_text_correct)
binary_ciphertext_2=convertion(cipher_text_incorrect)
binary_plaintext=convertion(plaintext)
```

**Spliting each bit of the binary number**

```python
def spliting_values(text):
  row = []
  for i in range(len(text)):
    if len(text[i]) < 8:
      pass
    else:
      for j in range(0,8):
        row.append(text[i][j])
  return row

cipher_row_1 = spliting_values(binary_ciphertext_1)
cipher_row_2 = spliting_values(binary_ciphertext_2)
plain_row = spliting_values(binary_plaintext)
```

**Creating DataFrame**

```python
df=pd.DataFrame([], columns =list(range(1,66)))
```

```python
loop=math.ceil(len(cipher_row_1) / 32)
m=0
n=32
label1=[1]
label0=[0]
for i in range(loop):
  a = plain_row[m:n] + cipher_row_1[m:n]+label1
  b = plain_row[m:n] + cipher_row_2[m:n]+label0
  if len(a) < 64:
    pass
  else:
    df.loc[len(df)] = a
    df.loc[len(df)] = b
  m+=32
  n +=32
```
